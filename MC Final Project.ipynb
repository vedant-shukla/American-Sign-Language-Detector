```python
import pandas as pd
import numpy as np
import glob
from sklearn.decomposition import PCA
```

```python
path = 'D:\Studies\ASU\Spring 18\CSE 535\group30\AllCombined\*.csv'

#calculating the fft for each row of the file: 

for fname in glob.glob(path):
    print(fname)
    data = pd.read_csv(fname,header=None)
    print(data.shape)
    
    #dropping the first column of ALX,ALY

    data = data.iloc[0:,1:]
        
    print(data.shape)
    arr = []
    
    i = 0
    while i < len(data):
        row = data.iloc[i,:]
        row = np.array(row)
        so = []
        so = np.fft.fft(row.astype(float))
        so = np.array(so)
        i = i+1
        arr.append(so)
    
    #storing the fft transformed into a dataframe
    
    print(np.shape(arr))
    converted_data = pd.DataFrame(arr)
    #storing the dataframe in a different folder:
    filename_whole = fname.split("\\")
    filename = filename_whole[7]
    print(filename)
    filepath = 'D:\Studies\ASU\Spring 18\CSE 535\group30\FFT_Converted\'' +filename+ '.csv'
    print(filepath)
    print(converted_data.shape)
    
    converted_data.to_csv(filepath)
    
    
```

```python
#applying the Knn Classifier:

#first we will randomly shuffle the segments:
data = pd.read_csv("Final_data.csv",header = None)

print("SHape: ",data.shape)
data = np.array(data)
print("Data: ",data[1])
np.random.shuffle(data)
print("New: ",np.shape(data))
print("New data",data[1])


```

```python

#separating x and y labels:
X = data[:,:-1]
Y = data[:,236]

#importing train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size = 0.2,random_state = 0)

#importing Knn classifier:
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(xtrain, ytrain)



```

```python
#predicting values based on Knn algo
ans = knn.predict(xtest)
print(np.shape(ans))
print(np.shape(ytest))
```

```python
sum = 0
for i in range(len(ans)):
    if(ytest[i] == ans[i]):
        sum = sum +1

print("Accuracy of KNN: ",sum/(len(ytest))*100)
```

```python
#applying naive bayes to our dataset

from sklearn.naive_bayes import GaussianNB
gausian_naive_bayes = GaussianNB()
gausian_naive_bayes.fit(xtrain, ytrain)
```

```python
#calculating the accuracy for naive bayes classifier:

ans = gausian_naive_bayes.predict(xtest)
sum = 0
for i in range(len(ans)):
    if(ytest[i] == ans[i]):
        sum = sum +1

print("Accuracy of Naive Bayes: ",sum/(len(ytest))*100)
```
